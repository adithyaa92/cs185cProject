```python
import pandas as pd

# Load the data set
df = pd.read_csv("nfl_draft.csv")

# Separate data by the name of university
# Count the rows in each university
uni_count = df.groupby(['College/Univ']).size().reset_index(name='counts')

# Sort the data by counts
uni_count_sorted = uni_count.sort_values(['counts'], ascending=False)

# Only take top 20 university
uni_count_sorted_20 = uni_count_sorted[0:20]

# Only take top 5 university
uni_count_sorted_5 = uni_count_sorted[0:5]

# Output first 20 unviersity
uni_count_sorted_20.head(20)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

# Get the evenly spaced value
# 0, 1, 2, 3,..., Length of uni_count_sorted_5
y_pos = np.arange(len(uni_count_sorted_5))

# Take only counts column 
count = uni_count_sorted_5['counts']
 
# bar(x, height, *, align='center', **kwargs)
plt.bar(y_pos, count, align='center', alpha=0.5)

# Set the label of each bar
plt.xticks(y_pos, uni_count_sorted_5['College/Univ'])

# Set the label of Y
plt.ylabel('Players Drafted')

# Set the label barchart
plt.title('NFL Draft')
 
# Output the barchart
plt.show()
```

```python
# Florida st. number of draft pick per year

# Only select data of Florida St.
df_florida_st = df.loc[df['College/Univ'] == 'Florida St.']

# Separate the data by year
# And count the rows by year
uni_count_florida_st = df_florida_st.groupby(['Year']).size().reset_index(name='counts')

# Sort the data by Year
uni_count_florida_st_sort = uni_count_florida_st.sort_values(['Year'], ascending=True)

# Y: Number of player picked per year
# X is Year
plt.plot(uni_count_florida_st_sort['Year'], uni_count_florida_st_sort['counts'])

# Output the line chart
plt.show()
```

```python
# Select top five college/university
# and store them into separate dataframes
df_florida_st = df.loc[df['College/Univ'] == 'Florida St.']
df_ohio_st = df.loc[df['College/Univ'] == 'Ohio St.']
df_miami_fl = df.loc[df['College/Univ'] == 'Miami (FL)']
df_florida = df.loc[df['College/Univ'] == 'Florida']
df_usc = df.loc[df['College/Univ'] == 'USC']

# Get the number of player picked per year
uni_count_florida_st = df_florida_st.groupby(['Year']).size().reset_index(name='counts')
uni_count_ohio_st = df_ohio_st.groupby(['Year']).size().reset_index(name='counts')
uni_count_miami_fl = df_miami_fl.groupby(['Year']).size().reset_index(name='counts')
uni_count_florida = df_florida.groupby(['Year']).size().reset_index(name='counts')
uni_count_usc = df_usc.groupby(['Year']).size().reset_index(name='counts')

# Sort by year
uni_count_florida_st_sort = uni_count_florida_st.sort_values(['Year'], ascending=True)
uni_count_ohio_st_sort = uni_count_ohio_st.sort_values(['Year'], ascending=True)
uni_count_miami_fl_sort = uni_count_miami_fl.sort_values(['Year'], ascending=True)
uni_count_florida_sort = uni_count_florida.sort_values(['Year'], ascending=True)
uni_count_usc_sort = uni_count_usc.sort_values(['Year'], ascending=True)

# Y: Number of player picked per year
# X: Year
plt.plot(uni_count_florida_st_sort['Year'], uni_count_florida_st_sort['counts'])
plt.plot(uni_count_ohio_st_sort['Year'], uni_count_ohio_st_sort['counts'])
plt.plot(uni_count_miami_fl_sort['Year'], uni_count_miami_fl_sort['counts'])
plt.plot(uni_count_florida_sort['Year'], uni_count_florida_sort['counts'])
plt.plot(uni_count_usc_sort['Year'], uni_count_usc_sort['counts'])

# Show the legen on top left
plt.legend(['Florida St','Ohio St', 'Miami (FL)', 'Florida', 'USC'], loc='upper left')

# Show the line chart
plt.show()
```
